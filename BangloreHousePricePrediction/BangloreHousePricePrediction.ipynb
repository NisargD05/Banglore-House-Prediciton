```python
def convert_sqft_to_num(x):
    if isinstance(x, str):
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        try:
            return float(x)
        except:
            return None
    return x
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)
```

```python
df['total_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price']*100000/ df['total_sqft']
```

```python
df['price_per_sqft']
```

```python
df.head()
```

```python
# Remove leading/trailing spaces from location names
df['location'] = df['location'].apply(lambda x: x.strip())

# Count how many times each location appears
location_count = df['location'].value_counts()

# Find locations that occur 10 times or less
location_count_less_10 = location_count[location_count <= 10]

# Replace rare locations with 'other'
df['location'] = df['location'].apply(
    lambda x: 'other' if x in location_count_less_10 else x
)
```

```python
df = df[(df['total_sqft'] / df['bhk']) >= 300]
```

```python
df.price_per_sqft.describe()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        
        low = m - 2*st
        high = m + 2*st
        
        gen_df = subdf[
            (subdf.price_per_sqft >= low) &
            (subdf.price_per_sqft <= high)
        ]
        
        df_output = pd.concat([df_output, gen_df], ignore_index=True)
    
    return df_output
```

```python
df = remove_outliers_sqft(df)
```

```python
df.drop(columns='price_per_sqft',inplace=True)
```

# NOW ML

```python
df = pd.get_dummies(df, columns=['location'], drop_first=True)
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_test_scaled)
print("Linear R2:", r2_score(y_test, y_pred_lr))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

param_grid = {
    'alpha': [0.01, 0.1, 1, 10, 50, 100]
}

ridge_cv = GridSearchCV(
    ridge,
    param_grid,
    scoring='r2',
    cv=5
)

ridge_cv.fit(X_train_scaled, y_train)
```

```python
print("Best alpha (Ridge):", ridge_cv.best_params_)
print("Best CV R2:", ridge_cv.best_score_)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(max_iter=10000)

param_grid = {
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1]
}

lasso_cv = GridSearchCV(
    lasso,
    param_grid,
    scoring='r2',
    cv=5
)

lasso_cv.fit(X_train_scaled, y_train)
```

```python
print("Best alpha (Lasso):", lasso_cv.best_params_)
print("Best CV R2:", lasso_cv.best_score_)
```

```python
best_ridge = ridge_cv.best_estimator_
best_lasso = lasso_cv.best_estimator_

print("Ridge Test R2:", best_ridge.score(X_test_scaled, y_test))
print("Lasso Test R2:", best_lasso.score(X_test_scaled, y_test))
```
